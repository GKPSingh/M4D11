```python
x = np.linspace(0,10,100)
x
```

```python
y = x**2
fig, ax = plt.subplots(figsize = (10,8))
ax.plot(x, y)
ax.set(title = 'Graph', ylabel = 'y = x^^2', xlabel = 'x');
```

## Scatter Plot

```python
y = np.exp(x)
fig, ax = plt.subplots(figsize = (10,8))
ax.scatter(x,y)
ax.set(Title = 'Graph', ylabel = 'y = exp(x)', xlabel = 'x');
```

```python
# Another Scatter plot
fig, ax = plt.subplots(figsize = (10,8))
ax.scatter(x, np.sin(x))
ax.set(Title = 'Amazing Sine Wave', ylabel = 'y = sin(x)', xlabel = 'x');
```

```python
# make a plot from dictionary
nut_butter_prices = {'Almond Butter':10, 'Peanut Butter':15, 'Cashew Butter':12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(Title = 'Gurkirpal Butter Shop', xlabel = 'Type of Butter', ylabel = 'Price of Butter ($)');
```

```python
nut_butter_prices = {'Almond Butter':10, 'Peanut Butter':15, 'Cashew Butter':12}
fig, ax = plt.subplots(figsize = (10, 8))
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
ax.set(Title = 'Gurkirpal Butter Shop', ylabel = 'Type of Butter', xlabel = 'Price of Butter ($)');
```

```python
## Making Histogram
x  = np.random.randn(1000)
fig,ax =plt.subplots()
ax.hist(x);
ax.set(Title = 'HISTOGRAM', ylabel = 'Frequency', xlabel = 'Values');
```

## Making Subplots

```python
fig, ((ax1, ax2), (ax3, ax4))= plt.subplots(nrows = 2, ncols=2, figsize = (10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.randn(10),np.random.randn(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

### Other way to plot subplots

```python
fig, ax = plt.subplots(nrows = 2, ncols =2, figsize = (10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.randn(10),np.random.randn(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

## Plotting from Pandas DataFrame

```python
car_sales = pd.read_csv('car-sales.csv')
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\, \.]', '')
car_sales
```

```python
car_sales.dtypes
```

```python
car_sales['Price'] = car_sales['Price'].str[:-2]
car_sales
```

```python
car_sales['Sale Date'] = pd.date_range('1/1/2021', periods = len(car_sales))
```

```python
car_sales
```

```python
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

car_sales.plot('Sale Date', 'Total Sales');

```python
car_sales['Price'] = car_sales['Price'].astype(int) 
car_sales.plot('Odometer (KM)', 'Price', kind = 'scatter')
```

```python
# Pyplot Method
over_50.plot(kind = 'scatter', x= 'age', y = 'chol', c='target');
```

```python
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter', y ='chol', x = 'age', c= 'target', ax =ax);
```

```python
## OO method mixed with pyplot method
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter', y ='chol', x = 'age', c= 'target', ax =ax);
ax.set_xlim([45,100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])
ax.set(Title = 'Heart Disease and Cholestrol Level', xlabel = 'Age', ylabel = 'Cholestrol');
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])
ax.set(Title = 'Heart Disease and Cholestrol Level', xlabel = 'Age', ylabel = 'Cholestrol');
ax.legend(*scatter.legend_elements(), title = 'Target');
```

```python
# Customize the plot
fig, ax = plt.subplots(figsize = (10, 6))
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])
ax.set(Title = 'Heart Disease and Cholestrol Level', xlabel = 'Age', ylabel = 'Cholestrol');
# Add a legend
ax.legend(*scatter.legend_elements(), title = 'Target');
# Add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle ='--' );
```

```python
# Subplot of chal, thalach
fig,(ax0, ax1) = plt.subplots(nrows = 2, ncols =1, figsize = (10,10))
# Add data tp ax0
scatter = ax0.scatter(x = over_50['age'], y = over_50['chol'], c= over_50['target'])
# Customize ax0
ax0.set(Title = 'Heart Disease and Cholestrol Levels', xlabel = 'Age', ylabel = 'Cholestrol')
# Add a legend
ax0.legend(*scatter.legend_elements(), title = 'Target')
# Add a meanline
ax0.axhline(y=over_50['chol'].mean(), linestyle='--' );



# Add data tp ax0
scatter = ax1.scatter(x = over_50['age'], y = over_50['thalach'], c= over_50['target'])
# Customize ax0
ax1.set(Title = 'Heart Disease and Cholestrol Levels', xlabel = 'Age', ylabel = 'Thalach')
# Add a legend
ax1.legend(*scatter.legend_elements(), title = 'Target')
# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--' );
```

```python
# Subplot of chal, thalach
fig,(ax0, ax1) = plt.subplots(nrows = 2, ncols =1, figsize = (10,10))
# Add data tp ax0
scatter = ax0.scatter(x = over_50['age'], y = over_50['chol'], c= over_50['target'])
# Customize ax0
ax0.set(Title = 'Heart Disease and Cholestrol Levels', xlabel = 'Age', ylabel = 'Cholestrol')
# Add a legend
ax0.legend(*scatter.legend_elements(), title = 'Target')
# Add a meanline
ax0.axhline(y=over_50['chol'].mean(), linestyle='--' );



# Add data tp ax0
scatter = ax1.scatter(x = over_50['age'], y = over_50['thalach'], c= over_50['target'])
# Customize ax0
ax1.set(Title = 'Heart Disease and Cholestrol Levels', xlabel = 'Age', ylabel = 'Thalach')
# Add a legend
ax1.legend(*scatter.legend_elements(), title = 'Target')
# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--' );

# Add atitle to the figure
fig.suptitle('Heart_Disease Analyisis', fontsize =16, fontweight = 'bold');
```
