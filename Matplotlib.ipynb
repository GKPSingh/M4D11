```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3,4,5]);
```

```python
x = [1,2,3,4,5]
y= [11,12,13,14,15]
plt.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.plot(x,y);
```

```python
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(x,y);
ax.set(title = 'Simple Plot', xlabel = 'x-axis', ylabel = 'y-axis');
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = x**2
fig, ax = plt.subplots(figsize = (10,8))
ax.plot(x, y)
ax.set(title = 'Graph', ylabel = 'y = x^^2', xlabel = 'x');
```

## Scatter Plot

```python
y = np.exp(x)
fig, ax = plt.subplots(figsize = (10,8))
ax.scatter(x,y)
ax.set(Title = 'Graph', ylabel = 'y = exp(x)', xlabel = 'x');
```

```python
# Another Scatter plot
fig, ax = plt.subplots(figsize = (10,8))
ax.scatter(x, np.sin(x))
ax.set(Title = 'Amazing Sine Wave', ylabel = 'y = sin(x)', xlabel = 'x');
```

```python
# make a plot from dictionary
nut_butter_prices = {'Almond Butter':10, 'Peanut Butter':15, 'Cashew Butter':12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(Title = 'Gurkirpal Butter Shop', xlabel = 'Type of Butter', ylabel = 'Price of Butter ($)');
```
