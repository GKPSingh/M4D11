```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3,4,5]);
```

```python
x = [1,2,3,4,5]
y= [11,12,13,14,15]
plt.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.plot(x,y);
```

```python
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(x,y);
ax.set(title = 'Simple Plot', xlabel = 'x-axis', ylabel = 'y-axis');
```
